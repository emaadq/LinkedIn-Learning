What is Machine Learning?

Giving random inputs and x outputs - can computer figure out what operations using the numbers will yield in the results?
3 types of learning: supervised, unsupervised, and reinforced learning

What is not Machine Learning?
ML draws from statistical concepts, but it encompasses much more than this. Goals of ML differ from goals of basic statistical modelling.
ML is concerned with results in the form of predicitons; predict future outcomes based on past events. Statistical models are concerned with the relationship between variables (inference); we want to know what happens to x from a change in y.
Statistical Learning is a term used to encapsulate the both.
ML prioritizes prediction, whereas Data Mining covers previously unknown patterns in data.
Machine Learning can also be referred to as Predictive Analytics: predict future outcomes
Data Mining as Descriptive Analytics: identify patterns in past data
Optimization as Prescriptive Analytics: to get a recommendation on how to proceed

Deep Learning is under the umbrella of Supervised learning.
All ML is AI, but not all AI is ML.

Artificial Intelligence Models can be categorized as:
1) Discriminative AI - models that focus on categorizing input data or on predicting a future outcome based on historical data; when looking at reviews for a restaraunt, discriminative AI can categorize into bad, good, or great.
*The learnings from this course are going to focus on Discriminative AI 
2) Generative AI - models designed to create NEW content based on user input. LLMs are a specific type of gen AI models that are good at replicating human text

2) Also have demographic information about each customer (age, gender, income level, etc.)
We pass this data into an unsupervised machine learning model (it is called unsupervised because there is no pre-existing rubric on how to evaluate the output)

What is Supervised Learning?

Supervised Learning -> the process of training a predictive model
Predictive models are used to assign labels to unlabeled data based on patterns laerned from previously labeleled historical data (we are using the data we have to assign the data we are getting); useful for predicting the outcome of a new event.
Ex: we are working in the analytics department in a credit union. We want to create a model that is going to predict loan risk; is a customer going to default on a loan or not. We have the following data:
1) Descriptive data about previous loans: loan amount, grade, income and purpose
2) Outcome of each previously issued loan
Before we can use the model to predict, we need to train it on the data we have. Input (Independent vairable) and Output (Dependent variable).

We would then give the model 3 independent variables, and the dependent variable; this will allow the model to connect the input to the output. Once we train it, we can evaluate how well it is performing on our results by just providing input values (without the output values). We then compare the outputs with 1) only inputs and 2) inputs and outputs.
This process is called: Predictive Accuracy of the model

Experience: E (training data)
Class of tasks: T (predict who will default)
Performance measure: P (predictive accuracy)
Spinning off Tom Mitchell's quote on ML models:
"A loan prediction model is said to learn if its ability to predict which borrowers will default on loans T, as measured by predictive accuracy P, improves as it encounters more training data E."

What is Reinforcement Learning?

Reinforcement learning: the science of learning to make decisions from interactions (applied in autonomous driving and gaming)

1) Finds previously unknown solutions to a problem (Ex: machine that plays chess better than any human)
2) Finding solutions to problems that arise due to unforseen circumstances (Ex: machine that is able to find a detour route as the primary route had bad weather/structure fell down)
Agent -> interacts -> Environment
Environment -> feedback -> Agent
Feedback comes back in (State, Reward)

Take Tik-Tak=Toe:

In the first sequence of the game, all the moves are going to have the same value associated with them; 0.5. But once we run the game one time and see a player win/the sequences that led to it, previous moves that setup the checkmate for TTT will now have a higher priority because the model learned from previous runthroughs that x input leads to y output. This is essentially how we train the model on data.

Need a balanced approach of these 2 to maximize learning:
Explotation: choosing the action that maximizes reward (can only solve problems previously encountered)
Exploration: choosing an action with no consideration of the reward (this allows for new patterns/formulas to be discovered)

What are the steps to Machine Learning?

1) Data Collection: identify and acquire the data we need for Machine Learning. For unsupervised learning: this is the unlabeled data with unknown patterns we are going to discover. For supervised learning: labeled historical data we are going to use to train our model. Refinforced learning: actions that yield the most reward (gathering ingredients for the salad)
2) Data Exploration: process of describing, visualizing and analyzing data in order to better understand it. Answers question such as how many rows/columns are within the data, are the duplicate values or outlires in the data? (inspecting ingredients to make sure they are not bad)
3) Data Preperation: process of making sure that our data is suitable for the ML we are going to use. Also includes modifying/transforming the data so it becomes easier to work with (normalizing it or reducing the rows/columns we need to work with). In the salad example, this includes chopping vegetables to get the desired salad we want (chopped/diced vegetables, etc.)
*80% of time in ML is spent on getting the data
4) Modeling: applying a ML approach to our data; applying the right type of ML model to our data to get the desired outcome. (Mixing ingredients we previously prepared, we need to mix more/less of x ingredients and decide on which to include/not include)
5) Evaluation: assess how well the ML approach we chose work. In unsupervised learning, we can evaluate by measuring how well it did and provide labels to categorize them. (Taste test the salad, add salt/pepper if need more seasoning or more dressing if it is dry)
6) Actionable Insight: identifying what to do based on the results of the ML approach we chose. Do we deploy our model to production? (Do we serve the salad or not?)
